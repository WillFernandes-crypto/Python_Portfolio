```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pickle

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from yellowbrick.classifier import ConfusionMatrix
from sklearn.naive_bayes import GaussianNB
```

```python
with open('../data/census.pkl', 'rb') as file:
    X_census_train, y_census_train, X_census_test, y_census_test = pickle.load(file)
```

```python
X_census_train.shape, y_census_train.shape, X_census_test.shape, y_census_test.shape
```

```python
naive_census = GaussianNB()
naive_census.fit(X_census_train, y_census_train)
```

```python
pred_census = naive_census.predict(X_census_test)
pred_census
```

```python
accuracy_score(pred_census, y_census_test)
```

```python
cm = ConfusionMatrix(naive_census)
cm.fit(X_census_train, y_census_train)
cm.score(X_census_test, y_census_test)
```

```python
print(classification_report(y_census_test, pred_census))
```
