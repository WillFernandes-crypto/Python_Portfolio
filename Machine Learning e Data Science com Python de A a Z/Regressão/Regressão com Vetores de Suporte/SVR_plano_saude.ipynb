```python
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
base = pd.read_csv('../data/plano_saude2.csv')
base
```

```python
X = base.iloc[:, 0].values
y = base.iloc[:, 1].values
```

```python
# linear kernel
regressor_linear = SVR(kernel='linear')
regressor_linear.fit(X.reshape(-1, 1), y)
```

```python
graphic1 = px.scatter(x=X, y=y)
graphic1.add_scatter(x=X, y=regressor_linear.predict(X.reshape(-1, 1)), name='Regression')
graphic1.show()
```

```python
# polynomial kernel
regressor_poly = SVR(kernel='poly', degree=3)
regressor_poly.fit(X.reshape(-1, 1), y)
```

```python
graphic2 = px.scatter(x=X, y=y)
graphic2.add_scatter(x=X, y=regressor_poly.predict(X.reshape(-1, 1)), name='Regression')
graphic2.show()
```

```python
scaler_x = StandardScaler()
X_scaled = scaler_x.fit_transform(X.reshape(-1, 1))
scaler_y = StandardScaler()
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))
```

```python
# rbf kernel
regressor_rbf = SVR(kernel='rbf')
regressor_rbf.fit(X_scaled, y_scaled)
```

```python
graphic3 = px.scatter(x=X_scaled.ravel(), y=y_scaled.ravel())
graphic3.add_scatter(x=X_scaled.ravel(), y=regressor_rbf.predict(X_scaled), name='Regression')
graphic3.show()
```
