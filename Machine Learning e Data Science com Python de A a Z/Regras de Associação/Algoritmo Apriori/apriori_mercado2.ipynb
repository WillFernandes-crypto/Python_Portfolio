```python
import pandas as pd

from apyori import apriori
```

```python
base = pd.read_csv('../data/mercado2.csv', header=None)
base
```

```python
transactions = []

for i in range(base.shape[0]):
    transactions.append([str(base.values[i, j]) for j in range(base.shape[1])])
```

```python
rules = apriori(transactions, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2)
results = list(rules)
len(results)
```

```python
A = []
B = []
support = []
confidence = []
lift = []

for result in results:
    s = result[1]
    result_rules = result[2]
    for result_rule in result_rules:
        A.append(result_rule[0])
        B.append(result_rule[1])
        support.append(s)
        confidence.append(result_rule[2])
        lift.append(result_rule[3])
```

```python
rules_df = pd.DataFrame({'A': A, 'B': B, 'support': support, 'confidence': confidence, 'lift': lift})
rules_df.sort_values(by='lift', ascending=False)
```
