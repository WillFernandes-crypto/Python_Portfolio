```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
```

```python
base = pd.read_csv('../data/credit_card_clients.csv', header=1)
base
```

```python
base['BILL_TOTAL'] = base['BILL_AMT1'] + base['BILL_AMT2'] + base['BILL_AMT3'] + base['BILL_AMT4'] + base['BILL_AMT5'] + base['BILL_AMT6']
```

```python
base.columns
```

```python
X = base.iloc[:, [1, 2, 3, 4, 5, 25]].values
X
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
graphic = px.line(x=range(1, 11), y=wcss)
graphic.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
labels = kmeans.fit_predict(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
X_pca.shape
```

```python
graphic = px.scatter(x=X_pca[:, 0], y=X_pca[:, 1], color=labels)
graphic.show()
```

```python
clients_list = np.column_stack((base, labels))
clients_list = clients_list[clients_list[:, 26].argsort()]

clients_list
```
