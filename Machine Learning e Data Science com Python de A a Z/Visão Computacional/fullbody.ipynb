```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('imgs/pessoas.jpg')
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Image')
plt.axis('off')
plt.show()

'''cv2.namedWindow('Image', cv2.WINDOW_NORMAL)
cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()'''
```

```python
detector = cv2.CascadeClassifier('cascades/fullbody.xml')
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.title('Image Gray')
plt.axis('off')
plt.show()

'''cv2.namedWindow('Image', cv2.WINDOW_NORMAL)
cv2.imshow('Image', gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()'''
```

```python
detections = detector.detectMultiScale(gray_img, scaleFactor=1.03, minNeighbors=5, minSize=(40, 40), flags=cv2.CASCADE_SCALE_IMAGE)
```

```python
detections
```

```python
len(detections)
```

```python
for (x, y, w, h) in detections:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Image')
plt.axis('off')
plt.show()
```
